Save your results as a pandas DataFrame called top_10_schools containing the "school_name" and a new column named "total_SAT", with results ordered by "total_SAT" in descending order ("total_SAT" being the sum of math, reading, and writing scores).

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

## 1. Finding schools with the best math scores
Subset the data to find the schools with math scores of at least 80%, considering the maximum possible score is 800 and save this as a pandas DataFrame called best_math_schools.

* The threshold should be 80% of the maximum possible score of 800.

* You can use boolean indexing to filter the data for schools meeting this threshold for "average_math", then subset the columns you need.

* Boolean indexing with a DataFrame uses square brackets syntax like this: df[df["column_name"] < 10]

* You can chain .sort_values() to sort the results, passing your column of interest.

* To sort from largest to smallest, remember to set ascending=False.

```python
# Which schools are best for math?
best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)

best_math_schools
```

## 2. Identifying the top 10 performing schools

Find the 10 best performing schools based on the total score across the three SAT sections.

Creating a column for average scores:
* Add a new column called "total_SAT" to the original DataFrame, which is the sum of the three existing SAT columns (math, reading, and writing).

Sorting and limiting the data
* Sort the values of the DataFrame by "total_SAT" in descending order.

* Subset the two columns you need.

* Limit the results to 10 rows with .head(10).

```python
# Calculate total_SAT per school
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# Who are the top 10 performing schools?
top_10_schools = schools.sort_values("total_SAT", ascending=False)[["school_name", "total_SAT"]].head(10)

top_10_schools
```

## 3. Locating the NYC borough with the largest standard deviation in SAT performance

Find out the number of schools, average SAT, and standard deviation of SAT for the NYC borough with the largest standard deviation, rounded to two decimal places.

Grouping the data by borough
* Group the data by "borough" and find the count of schools, mean and standard deviation of "total_SAT".

* You can chain .groupby() with .agg() to calculate these values.

* The .agg() method accepts a list of statistics, as strings, to calculate, e.g., .agg(["min", "max", "median"]).

* Chain .round() to round the data to a specified number of decimal places.


Filtering for the largest standard deviation
* Create a new Dataframe called large_std_dev that subsets the boroughs DataFrame for the row where "std" is equal to the .max() value for that column.


Renaming columns
* To rename multiple columns you can use the df.rename() method, passing a dictionary to the columns keyword argument; the original column names are the keys and the new column names ("num_schools", "average_SAT", "std_SAT") are the values.

```python
# Which NYC borough has the highest standard deviation for total_SAT?
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

boroughs
```

```python
# Filter for max std and make borough a column
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]

# Rename the columns for clarity
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})

# Optional: Move borough from index to column
largest_std_dev.reset_index(inplace=True)

largest_std_dev
```
